```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/abc.jpeg', cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if image is None:
    print("Error: Could not load image.")
    exit()

# Define Prewitt kernels
prewitt_x = np.array([[ -1, 0, 1],
                       [ -1, 0, 1],
                       [ -1, 0, 1]])

prewitt_y = np.array([[ 1,  1,  1],
                       [ 0,  0,  0],
                       [-1, -1, -1]])

# Apply the Prewitt operator
gradient_x = cv2.filter2D(image, cv2.CV_64F, prewitt_x)
gradient_y = cv2.filter2D(image, cv2.CV_64F, prewitt_y)

# Compute the magnitude of the gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
gradient_magnitude = np.uint8(np.clip(gradient_magnitude, 0, 255))  # Convert to uint8

# Display the results
plt.figure(figsize=(10, 8))
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Prewitt X Gradient')
plt.imshow(gradient_x, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Prewitt Y Gradient')
plt.imshow(gradient_y, cmap='gray')
plt.axis('off')

plt.figure(figsize=(5, 5))
plt.title('Gradient Magnitude')
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')

plt.show()
```
